```python
import pandas as pd
import matplotlib.pyplot as plt

# Deixa os números mais legíveis
pd.options.display.float_format = '{:,.2f}'.format

urls = {
    'Loja 1': 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv',
    'Loja 2': 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv',
    'Loja 3': 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv',
    'Loja 4': 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv'
}

lojas = {nome: pd.read_csv(url) for nome, url in urls.items()}
```

# **Vendas por Categoria e Produto**

```python
for nome, df in lojas.items():
    print(f"\n=== {nome} ===")

    print("\nVendas por Categoria:")
    print(df['Categoria do Produto'].value_counts())

    print("\nTop 5 Produtos Mais Vendidos:")
    print(df['Produto'].value_counts().head(5))

    # Gráfico de barras para categorias
    plt.figure(figsize=(7,4))
    df['Categoria do Produto'].value_counts().plot(kind='bar', color='teal')
    plt.title(f'Vendas por Categoria – {nome}')
    plt.xlabel('Categoria')
    plt.ylabel('Quantidade de Vendas')
    plt.tight_layout()
    plt.show()

    # Gráfico de barras para top 5 produtos
    plt.figure(figsize=(7,4))
    df['Produto'].value_counts().head(5).plot(kind='bar', color='purple')
    plt.title(f'Top 5 Produtos – {nome}')
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.tight_layout()
    plt.show()
```

# **Média de Avaliações**

```python
avaliacoes = {
    nome: df['Avaliação da compra'].mean()
    for nome, df in lojas.items()
}

# Impressão
print("\n=== Média de Avaliações por Loja ===")
for nome, media in avaliacoes.items():
    print(f"{nome}: {media:.2f}")

# Gráfico
plt.figure(figsize=(7,5))
plt.bar(avaliacoes.keys(), avaliacoes.values(), color='orange')
plt.title('Média de Avaliações por Loja')
plt.ylabel('Nota Média')
plt.xlabel('Lojas')
plt.tight_layout()
plt.show()
```

# **Produtos Mais e Menos Vendidos**

```python
for nome, df in lojas.items():
    contagem = df['Produto'].value_counts()
    mais = contagem.idxmax()
    menos = contagem.idxmin()

    print(f"\n=== {nome} ===")
    print(f"Produto mais vendido: {mais} ({contagem.max()} vendas)")
    print(f"Produto menos vendido: {menos} ({contagem.min()} venda)")

    # Gráfico top 10
    plt.figure(figsize=(8,4))
    contagem.head(10).plot(kind='bar', color='green')
    plt.title(f'Top 10 Produtos Mais Vendidos – {nome}')
    plt.ylabel('Vendas')
    plt.xlabel('Produto')
    plt.tight_layout()
    plt.show()
```

# **Frete Médio por Loja**

```python
fretes = {
    nome: df['Frete'].mean()
    for nome, df in lojas.items()
}

# Impressão
print("\n=== Frete Médio por Loja ===")
for nome, valor in fretes.items():
    print(f"{nome}: R$ {valor:.2f}")

# Gráfico
plt.figure(figsize=(7,5))
plt.bar(fretes.keys(), fretes.values(), color='salmon')
plt.title('Frete Médio por Loja')
plt.ylabel('Frete Médio (R$)')
plt.xlabel('Lojas')
plt.tight_layout()
plt.show()
```

 # **Faturamento Total por Loja**

```python
faturamentos = {
    nome: df['Preço'].sum()
    for nome, df in lojas.items()
}

# Impressão
print("\n=== Faturamento Total por Loja ===")
for nome, valor in faturamentos.items():
    print(f"{nome}: R$ {valor:,.2f}")

# Gráfico
plt.figure(figsize=(7,5))
plt.bar(faturamentos.keys(), faturamentos.values(), color='steelblue')
plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento (R$)')
plt.xlabel('Lojas')
plt.tight_layout()
plt.show()
```
